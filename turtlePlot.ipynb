```python
from geo  import *
import matplotlib .pyplot as plt

x_axis_plot = [i for i in range(1, 11)]
areas = [area(radius) for radius in x_axis_plot]
perimeters = [perimeter(radius) for radius in x_axis_plot]


plt.plot(x_axis_plot, areas, label = "Areas")
plt.plot(x_axis_plot, perimeters, label = "Perimeters")

plt.xlabel("radius")
plt.ylabel("Measurement")
plt.title("Area and Perimeter of the Circles")
plt.legend()

plt.show()


```

```python
from geo  import *
import matplotlib .pyplot as plt

degrees  = [i for i in range(-760, 760)] #x_axis

sins = [sin(degree) for degree in degrees]
coss = [cos(degree) for degree in degrees]

plt.plot(degrees, sins, label = "Sin")
plt.plot(degrees, coss, label = "Cos")
plt.axvline(x=0, color = "red", linestyle ="-",  label = "x = 0" )
plt.xlabel("Degrees")
plt.ylabel("Measurements")
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)  # Generate 100 points from 0 to 2π
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("Angle (radians)")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```
